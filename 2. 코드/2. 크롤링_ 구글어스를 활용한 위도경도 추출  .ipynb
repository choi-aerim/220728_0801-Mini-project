```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#컬럼 생성: 위도, 경도

idx =0
  
for i in aerim_df[aerim_df['위도']== '0'].index:
    address = aerim_df.loc[i, '세부 주소_clean']
    try:
        aerim_df.loc[idx,'위도'] = geocoding(address)[0]
        aerim_df.loc[idx,'경도'] = geocoding(address)[1]
    except:
        pass
    idx += 1
aerim_df
```
